```python
increase_of_prices_g = increase_of_prices.groupby('period').agg({'compound_score': 'mean'}).reset_index()
increase_prices_home_g = increase_prices_home.groupby('period').agg({'compound_score': 'mean'}).reset_index()
increase_prices_housing_g = increase_prices_housing.groupby('period').agg({'compound_score': 'mean'}).reset_index()
increase_price_home_g = increase_price_home.groupby('period').agg({'compound_score': 'mean'}).reset_index()
increase_price_housing_g = increase_price_housing.groupby('period').agg({'compound_score': 'mean'}).reset_index()
rising_prices_home_g = rising_prices_home.groupby('period').agg({'compound_score': 'mean'}).reset_index()
rising_prices_housing_g = rising_prices_housing.groupby('period').agg({'compound_score': 'mean'}).reset_index()
rising_prices_g = rising_prices.groupby('period').agg({'compound_score': 'mean'}).reset_index()
rising_price_home_g = rising_price_home.groupby('period').agg({'compound_score': 'mean'}).reset_index()
rising_price_housing_g = rising_price_housing.groupby('period').agg({'compound_score': 'mean'}).reset_index()
```

```python
increase_of_prices_g.rename(columns = {'compound_score':'increase_of_prices_g'}, inplace = True)
increase_prices_home_g.rename(columns = {'compound_score':'increase_prices_home_g'}, inplace = True)
increase_prices_housing_g.rename(columns = {'compound_score':'increase_prices_housing_g'}, inplace = True)
increase_price_home_g.rename(columns = {'compound_score':'increase_price_home_g'}, inplace = True)
increase_price_housing_g.rename(columns = {'compound_score':'increase_price_housing_g'}, inplace = True)
rising_prices_home_g.rename(columns = {'compound_score':'rising_prices_home_g'}, inplace = True)
rising_prices_housing_g.rename(columns = {'compound_score':'rising_prices_housing_g'}, inplace = True)
rising_prices_g.rename(columns = {'compound_score':'rising_prices_g'}, inplace = True)
rising_price_home_g.rename(columns = {'compound_score':'rising_price_home_g'}, inplace = True)
rising_price_housing_g.rename(columns = {'compound_score':'rising_price_housing_g'}, inplace = True)
```
