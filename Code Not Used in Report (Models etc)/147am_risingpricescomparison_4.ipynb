## Brown–Forsythe test on aggregated data

Generalization of the first F-test that was conducted on the 'rising prices' and 'rising price housing' datasets. This time, we use aggregate data to test if the conclusion we made about the significant difference in variance between 'rising prices' and 'rising price housing' is true in general, when taking in account all terms about the rise in housing prices and the rise in prices. This is a better option than doing multiple F-tests on pairs of datasets since we would have to conduct a lot of tests and it would almost certainly fail (becauce of high risk of type I error). Again, we choose a level of significance alpha = 0,05.

First, let's look at the distributions of the aggregated data.

```python
alt.Chart(house_price_g, title = 'Distribution of average compound sentiment score on aggregated data about the rise of housing prices').mark_bar().encode(
    alt.X('compound_score', bin=True),
    y='count()',
)
```

Both distributions don't look very normal when grouping by 'period'. But grouping by period, i.e by month is also more relevant because most datasets don't have the same number of tweets per day. Some datasets have only one tweet per day whereas others have a lot more of them. 
Since the distributions are not normally distribued, we are going to conduct a Brown-Forsythe Test, which is more robust than the F-test.

```python
# Brown-Forsythe Test
rise_price_np = rise_price_g['compound_score'].to_numpy()
house_price_np = house_price_g['compound_score'].to_numpy()
```

```python
var_rise_house = np.var(house_price_np)
var_rise_price = np.var(rise_price_np)

print(var_rise_house, var_rise_price)
```

```python
stats.levene(rise_price_np, house_price_np, center='median')
```

p_value < alpha = 0,05. Hence, we reject the null hypothesis. Average compound sentiment score on the increase/rise of housing prices has a significantly different standard deviation than average compound sentiment score on the increase/rise of prices. We can conclude that average sentiment towards the increase/rise of housing prices fluctuate more overtime than average sentiment towards the increase/rise of prices. Namely, people have more various reactions over time on topics related to the increase of prices on the housing market.
